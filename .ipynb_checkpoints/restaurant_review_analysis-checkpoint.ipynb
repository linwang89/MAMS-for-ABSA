```python
import pandas as pd
import xml.etree.ElementTree as ET
tree = ET.parse("data/MAMS-ACSA/raw/train.xml")
root = tree.getroot()
```

```python
categories = set()
for child in root:
    for i in child:
        if i.tag == 'aspectCategories':
            for j in i:
                if j.attrib['category'] not in categories:
                    categories.add(j.attrib['category'])
train_set = pd.DataFrame(columns=['sentence']+list(categories))
test_set = pd.DataFrame(columns=['sentence']+list(categories))
val_set = pd.DataFrame(columns=['sentence']+list(categories))
```

```python
count=0
for child in root:
    for i in child:
        if i.tag == 'text':
            train_set.loc[count,'sentence']=i.text
        if i.tag == 'aspectCategories':
            for j in i:
                train_set.loc[count,j.attrib['category']] = j.attrib['polarity']
    count += 1
```

```python
tree = ET.parse("data/MAMS-ACSA/raw/test.xml")
root = tree.getroot()
count=0
for child in root:
    for i in child:
        if i.tag == 'text':
            test_set.loc[count,'sentence']=i.text
        if i.tag == 'aspectCategories':
            for j in i:
                test_set.loc[count,j.attrib['category']] = j.attrib['polarity']
    count += 1
```

```python
tree = ET.parse("data/MAMS-ACSA/raw/val.xml")
root = tree.getroot()
count=0
for child in root:
    for i in child:
        if i.tag == 'text':
            val_set.loc[count,'sentence']=i.text
        if i.tag == 'aspectCategories':
            for j in i:
                val_set.loc[count,j.attrib['category']] = j.attrib['polarity']
    count += 1
```
